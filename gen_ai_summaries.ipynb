```python
import openai
from openai import OpenAI
from dotenv import load_dotenv
import os
load_dotenv()
```

```python
import fitz
```

```python
api_key = os.getenv("api_key")
```

```python
client = OpenAI(api_key=api_key)
```

```python
pdf_doc = fitz.open("Filings/Initial/2025-02-10/TNP65343.pdf" )
```

```python
# Extract text from all pages as a list
full_text_list = [page.get_text() for page in pdf_doc]
full_text = "\n".join(full_text_list)  # Join all pages into a single string
print(full_text)
```

```python
completion = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[
            {"role": "system", "content": "You are an analyst doing competitive intelligence research on insurance companies."},
            {"role": "user", "content": f"The following is a filing with a DOI from an insurance company: '{full_text}'. Give a summary of the rate filing."}
        ],
    max_tokens = 300
    #max_completion_tokens = 300
    )
print(completion)
```

```python
completion.choices[0].message.content
```

```python
anthropic_api_key = os.getenv("anthropic_api_key")
```

```python
import anthropic

client = anthropic.Anthropic(
    # defaults to os.environ.get("ANTHROPIC_API_KEY")
    api_key= anthropic_api_key  #"anthropic_api_key",
)
message = client.messages.create(
    model= "claude-3-haiku-20240307",
    max_tokens=1024,
    messages=[
        {"role": "user", "content": "Hello, Claude"}
    ]
)
print(message.content)
```
